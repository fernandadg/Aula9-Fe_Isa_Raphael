**MODELO PROBABILISTICO GAMA**

2.1.

```python
fig, ax = plt.subplots(1, 1)
a = 1.99
mean, var, skew, kurt = gamma.stats(a, moments='mvsk')
t = np.linspace(0,10,num=100)
ax.plot(t, gamma.pdf(t, a),'r-', lw=5, alpha=0.6, label='gamma pdf')
rv = gamma(a)
ax.plot(t, rv.pdf(t), 'k-', lw=2, label='frozen pdf')
vals = gamma.ppf([0.001, 0.5, 0.999], a)
np.allclose([0.001, 0.5, 0.999], gamma.cdf(vals, a))
True
r = gamma.rvs(a, size=1000)
ax.hist(r, normed=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.show()
```

2.2. A forma analítica da distribuição é dada por  $$f(x, alpha, r) = alpha/Gamma(r) * [(alpha*x)^{r-1}] * e^(-alpha*x)$$ 

2.3. Os parâmetros que descrevem essa distribuição são r (parâmetro de forma) e alpha (parâmetro de escala). 

2.4. A fórmula específica para calcular esperança (média) em função dos parâmetros é r/alpha e a da variância é r/(alphaˆ2).

2.5. Essa distribuição tem aplicações em modelagem de tempos de fila. Por exemplo tempo para que o último seja o primeiro. É uma boa distribuicao para usarmos nesse trabalho, já que estuda o tempo de abertura dos sites, de acordo com as pessoas que acessam. 

2.6. Essa distribuição é util para inferência estatística. Por exemplo, dada uma fila (amostra), a distribuição permite estudar características dessa, como o tempo para que uma pessoa chegue na primeira posição, em função de algum parâmetro de interesse.
